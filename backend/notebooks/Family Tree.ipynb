# Derive Versioning

Here is how versioning can be derived from the json, coded in Python.

```python
from json import dumps, loads
from datetime import datetime
from collections import OrderedDict
```

```python
ft_test = ft_input
print(dumps(ft_test, indent=2))
```

## Sort each lineage by timestamp

```python
ft_test
```

```python
def sort_lineages_by_timestamp(family_tree):
    '''
    Helper function to sort each lineage in a family tree by timestamp
    Input: Family tree dictionary
    Output: Sorted family tree OrderedDict
    '''
    sorted_family_tree = OrderedDict()

    for lineage_id, items in family_tree.items():
        # Convert the dictionary to a list of items
        items_list = [(key, details) for key, details in items.items()]
        # Sort items by timestamp
        sorted_items = sorted(items_list, key=lambda x: datetime.strptime(x[1]['timestamp'], '%Y-%m-%d %H:%M:%S'))
        # Convert back to an ordered dictionary
        sorted_family_tree[lineage_id] = OrderedDict(sorted_items)

    return sorted_family_tree
```

```python
# Sort each lineage by timestamp
ft_sorted = sort_lineages_by_timestamp(ft_test)
print(dumps(ft_sorted, indent=2))
```

## Add versioning to main lineage

```python
def find_main_lineage_id(data):
    '''
    Given a family_tree
    Finds lineage id of the main lineage (lineage with root dataset)
    '''
    for lineage_id, datasets in data.items():
        for dataset in datasets:
            if datasets[dataset]['derived_from'] == 'null':
                return lineage_id
    return None
```

```python
main_lineage_id = find_main_lineage_id(ft_sorted)
print("Main lineage id: ", main_lineage_id)
```

```python
# Used to store final output with versioning
ft_output = OrderedDict()
```

```python
print(ft_sorted[main_lineage_id])
```

```python
# Add versioning to main lineage
for i, key in enumerate(ft_sorted[main_lineage_id]):
    ft_sorted[main_lineage_id][key]['version'] = f"1.{i+1}"
# print("Main lineage with versions: ", dumps(ft_sorted[main_lineage_id], indent=2))

# Remove main lineage from ft_sorted, add to ft_output
ft_output[main_lineage_id] = ft_sorted[main_lineage_id]
if main_lineage_id in ft_sorted:
    del ft_sorted[main_lineage_id]
print("ft_output: ", dumps(ft_output, indent=2))
```

## Add versioning to other lineages

```python
# Helper function that checks if any items in a lineage contain a certain id
# Returns version number of that matched item
def contains_id(lineage, target_id):
    '''
    Checks if any datasets in a lineage contain a target_id
    Returns version number of that item with the target_id
    Returns None if no dataset matches the target_id
    '''
    print("*** Inside contains_id")
    print("Lineage: ", lineage)
    print("Target id: ", target_id)
    for dataset_id in lineage:
        print("Lineage: ", lineage)
        print("Lineage[dataset_id]: ", lineage[dataset_id])
        print("dataset_id: ", dataset_id)
        if dataset_id == target_id:
            return lineage[dataset_id]['version']
    return None
```

```python
def add_versions_to_lineage(lineage, root_version):
    '''
    Adds version numbers to a lineage based on the root_version number
    '''
    this_v = 0
    for i in lineage:
        lineage[i]['version'] = f'{root_version}.{this_v}'
        this_v+=1

    return lineage
```

```python
# Loop through sorted lineages, add lineages to output, until no lineages remaining
unfinished_lineage_ids = list(ft_sorted.keys())
print("Unfinished lineage ids: ", unfinished_lineage_ids)
print("Length: ", len(unfinished_lineage_ids))
print("ft_sorted: ", ft_sorted)
print("Length: ", len(ft_sorted))
while (len(unfinished_lineage_ids) > 0):
    print("*******")
    print("Unfinished lineage ids: ", unfinished_lineage_ids)
    # Loop through remaining sorted lineages, match with output lineages
    for lineage_sorted_id in ft_sorted:
        print("Lineage_sorted_id: ", lineage_sorted_id)

        # Get what node this lineage was derived from
        first_lineage_key, first_lineage_value = next(iter(ft_sorted[lineage_sorted_id].items()))
        print("First lineage key: ", first_lineage_key)
        print("First lineage value: ", first_lineage_value)
        this_derived_from = first_lineage_value['derived_from']
        print("This derived from: ", this_derived_from)

        # Match sorted lineage with output lineage
        for lineage_output_id in ft_output:
            print("Lineage output: ", lineage_output_id)
            print("ft_sorted: ", ft_sorted)
            root_version = contains_id(ft_output[lineage_output_id], this_derived_from)
            print("Root version: ", root_version)
            if root_version is not None:
                # Match found, create versions for this lineage
                lineage_sorted_versions = add_versions_to_lineage(ft_sorted[lineage_sorted_id], root_version)
                print("Lineage sorted versions: ", lineage_sorted_versions)

                # Add lineage to output
                ft_output[lineage_sorted_id] = lineage_sorted_versions

                # Remove lineage_id from unfinished_lineage_ids
                unfinished_lineage_ids.remove(lineage_sorted_id)
                
                break
```
